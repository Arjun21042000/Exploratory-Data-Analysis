```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.DataFrame({
    'Platform': ['Android', 'iOS', 'Windows', 'Android', 'iOS', 'Windows', 'Android', 'Windows', 'iOS', 'Windows'],
    'Likes': [100, 150, 200, 250, 300, 150, 200, 250, 300, 350]
})

Likes_by_platform = data.groupby('Platform')['Likes'].sum().reset_index()

# Plotting the pie chart
plt.figure(figsize=(8, 8))  # Adjust the size of the figure
plt.pie(likes_by_platform['Likes'], labels=likes_by_platform['Platform'], autopct='%1.1f%%', colors=sns.color_palette('rainbow'))
plt.title('Total Likes by Platform')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.DataFrame({
    'Platform': ['Android', 'iOS', 'Windows', 'Android', 'iOS', 'Windows', 'Android', 'Windows', 'iOS', 'Windows'],
    'Likes': [100, 150, 200, 250, 300, 150, 200, 250, 300, 350]
})
likes_by_platform = data.groupby('Platform')['Likes'].sum().reset_index()

colors = sns.color_palette('rainbow', len(likes_by_platform['Platform'].unique()))

color_map = dict(zip(likes_by_platform['Platform'].unique(), colors))

plt.figure(figsize=(10, 6)) 
bars = plt.bar(likes_by_platform['Platform'], likes_by_platform['Likes'], color=[color_map[platform] for platform in likes_by_platform['Platform']])

for bar, platform in zip(bars, likes_by_platform['Platform']):
    bar.set_color(color_map[platform])

plt.title('Total Likes by Platform')
plt.xlabel('Platform')
plt.ylabel('Total Likes')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



```
