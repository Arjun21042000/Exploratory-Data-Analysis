```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(data['Rating'].dropna(), kde=True, bins=20, color='Red')
plt.title("Distribution of App Rating")
plt.xlable('Rating')
plt.ylable('Frequency')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating', data=data, color='Yellow')
plt.title("Boxplot of App Rating")
plt.xlabel('Rating')
plt.show()
```
